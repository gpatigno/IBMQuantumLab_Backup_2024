#### <b> Preparación estatal
#### Se debe hacer una conversión de los datos clásicos a los de estado cuántico para poder trabajarlos en el espacio de Hilbert, atreves de un mapa de características cuánticas los métodos mas usados son: 
#### Incrustación de base: En este método, simplemente codificamos nuestros datos en cadenas binarias. Convertimos cada entrada a un estado de base computacional de un sistema qubit.
#### Incrustación de amplitud: Codificamos los datos como amplitudes de un estado cuántico, teniendo como referencia los N datos.
#### Incrustación de ángulo: Codificación de ángulo. Codificamos información clásica en rotaciones angulares de un qubit. 
#### <b> Circuito modelo
#### Se crea un operador unitario parametrizado, se construye a partir de puertas que evolucionan el estado de entrada, en pocas palabras es colocar estados en función de las entradas y evitar que puedan tomas todos los posibles resultados y restringirlos solo a los que son requeridos según el tipo de clasificación que se hace.
#### <b>Medición
#### Al final se realiza la medición en donde se calcula la probabilidad de un elemento que tanto corresponde a un grupo o subgrupo de su categorización.

### 1.4. ¿En qué consisten los mapas de características (Feature Maps), y cuáles son sus parámetros ajustables?
#### Es una conversión de un dato clásico a estados cuánticos que seria de mayor dimensión es básicamente una conversión vectorial pasando de una dimensión inferior a un superior, pasando del estado clásico al espacio de Hilbert.


#### Parameters
#### feature_dimension (int) – Número de caracteristicas

#### reps (int) – Número de circuitos repetidos. Tiene un valor por defecto de 2, y un valor minimo de 1.

#### data_map_func (Optional[Callable[[ndarray], float]])
#### Una funcion de mapeo para los datos en x que pueden ser suministrados para sobrecargar el mapeo del self_product().

#### parameter_prefix (str) – El prefijo es usado si los parametros por defecto son generados.

#### insert_barriers (bool) – Si es verdadero, se insertan barreras entre las etapas de evolución.

### 1.5. De acuerdo con lo visto en clase y estudiado en este artículo, explique el rol que desempeña la optimización en el entrenamiento de los algoritmos cuánticos de aprendizaje de máquina.
#### Se trabaja en el mejoramiento de los circuitos cuánticos que construyen un proceso de aprendizaje para disminuir el umbral de error que se pueda generar en los resultados de analizar los datos.
### 1.5.1. ¿Cuáles son los parámetros que son entrenados en los algoritmos cuánticos de aprendizaje de máquina descritos en el artículo?
#### **Adan:** Algoritmo de Estimación de Momento Adaptativo, es un proceso similar a un filtro de calman ya que funciona en base al error, a medida que va procesando la información, analiza el error de los resultados y ajusta el modelo para mejorarlo y disminuir el error con cada interacción.
#### **COBYLA:** Optimización Restringida por Aproximaciones Lineales, Construye aproximaciones lineales sucesivas de la función de pérdida y restringe a través de un simplex de n+1 puntos (en n dimensiones), y optimiza estas aproximaciones en una región de confianza en cada paso. COBYLA admite restricciones de igualdad transformándolas en dos restricciones de desigualdad.
#### **SPSA:** Es un proceso que se basa en mediciones de la función objetivo y trabaja los datos según lo requiere la salida para obtener lo mejores resultados y adaptarse al sistema, es similar al Adan pero este no se basa en el error si no en los resultados de la función resultado.
#### 1.5.2. ¿Cómo se realiza el entrenamiento de dichos parámetros, y cuál es el rol de la función MSE?
#### La funcion MSE es la medida de error que tiene el sistema con respecto al valor real que debio responder, hallar una MSE de cero es casi imposible ya que seria un sistema que responda a la perfeción cada caso.

```python
# Cargando simuladores y herramientas de Qiskit
from qiskit import BasicAer
from qiskit.utils import QuantumInstance, algorithm_globals
from qiskit.algorithms.optimizers import SPSA #Optimizador a utilizar
from qiskit.circuit.library import TwoLocal, ZZFeatureMap
from qiskit_machine_learning.algorithms import VQC
import qiskit
from qiskit import QuantumCircuit, execute, Aer
from qiskit.visualization import plot_histogram
from qiskit.quantum_info import Statevector

import warnings
warnings.filterwarnings("ignore")

# Herramientas de python para la lectura y manipulación de datos desde archivo
import numpy as np
import pandas as pd
import seaborn as sns # For pairplots and heatmaps
import matplotlib.pyplot as plt
from IPython.display import clear_output
from math import sqrt

# Herramientas para procesamiento de datos simbolicos y conversiones en los datasets
from sklearn.preprocessing import normalize
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.utils import shuffle

from sklearn.metrics import recall_score, precision_score, confusion_matrix
```

```python
# Numero de qubits/caracteristicas a analizar para entrenar el algoritmo
n = 7
RANDOM_STATE = 42
```

```python
def DataPreProcesing(DATA_PATH1 = "data/train.csv"):
    """
    Pre-Procesamiento de los datos almacenados en arreglos podados y normalizados
    """
    # Lectura de datos
    # En este paso se toman los datos crudos para manejarlos desde la memoria del programa
    train = pd.read_csv(DATA_PATH1) 
    # Preprocesamiento de los datos, descarte de datos inecesarios
    # En este paso tomamos los datos que se pueden considerar innecesarios para realizar el algoritmo,
    # además se completan datos faltantes.
    train = train.dropna(subset=["Embarked"])
    train = train.drop("Cabin", axis=1) 
    mean = train["Age"].mean()
    train["Age"] = train["Age"].fillna(mean)
    train = train.drop("PassengerId", axis=1)
    train = train.drop("Name", axis=1) 
    train = train.drop("Ticket", axis=1) 
    # Preparacion de datos Texto a Numericos
    # Se sabe que los computadores no procesan el texto, por esto se deben traducir los datos de texto a formas
    # que el computador pueda entender como lo son los números
    le = LabelEncoder()
    for col in ['Sex', 'Embarked']:
        le.fit(train[col])
        train[col] = le.transform(train[col])
    # Normalización de los datos
    # Luego de que todos los datos son numericos se realiza una normalización para 
    # obtener datos que oscilen entre 0 y 1 
    scaler = MinMaxScaler()
    scaler.fit(train)
    train = scaler.transform(train)    
    # Separación de los datos en subsets de entrenamiento y prueba.
    # Los datos se encuentran almacenados en arreglos columna donde la forma luego de aplicar los descartes
    # tiene como primera columna las etiquetas de sobrevivencia por tanto se toman las 7 restantes como datos de entrada
    input_data = train[:, 1:8] # para todos los pasajeros las columnas 1 a 8, datos independientes
    labels = train[:, 0] # para todos los pasajeros la columna 0(Sobrevivencia), variable de interés
    train_input, test_input, train_labels, test_labels = train_test_split(input_data, labels, test_size=0.2)
    return train_input, test_input, train_labels, test_labels
```

```python
X_train, X_test, Y_train, Y_test = DataPreProcesing()
```

```python
print('Se tienen {} filas de entrenamiento y {} de prueba'.format(X_train.shape[0], X_test.shape[0]))
print('Hay {} columnas de caracteristicas'.format(X_train.shape[1]))
```

```python
# Creacion de los circuitos entrelazados de caracteristicas
# Para cada característica de los pasajeros descrita se crea un qubit mediante el cual se entrelazará dicha característica con
# las demás tenidas en cuenta para el entrenamiento lo que facilitará al clasificador interpretar que las variables están 
# relacionadas sin la necesidad de conocer la forma deterministica de su relación, para realizar predicciones acerca de la variable de interés.
sv = Statevector.from_label('0' * n) # Estado inicial |0> para todos los qubits
feature_map = ZZFeatureMap(feature_dimension=n, reps=2, entanglement="linear") # Entrelazamiento lineal entre las caracteristicas y los qubits
display(feature_map.decompose().draw('mpl'))
```

```python
# Funcion de actualización
# Con cada iteración los valores de los pesos(influencia) a medida que cambie el valor de la función objetivo deberán ser 
# actualizados para realizar, ya que esta función es parametrizada depende de los parámetros (en este caso el valor de la función
# de costos) para realizar una nueva estimación de la funcion objetivo

ansatz = TwoLocal(feature_map.num_qubits, ['ry', 'rz'], 'cz', reps=2)
display(ansatz.decompose().draw('mpl'))
```

```python
# Funcion de graficación de datos obtenidos en tiempo real
def callback_graph(weights, obj_func_eval):
    clear_output(wait=True)
    objective_func_vals.append(obj_func_eval)
    plt.title("Objective function value against iteration")
    plt.xlabel("Iteration")
    plt.ylabel("Objective function value")
    plt.plot(range(len(objective_func_vals)), objective_func_vals)
    plt.show()
```

```python
# Instanciación del simulador
quantum_instance = QuantumInstance(Aer.get_backend("aer_simulator"), 
                                   shots=1024)
```

```python
# Variational Quantum Classifier(Clasificador Cuántico Variacional)
vqc = VQC(
    feature_map=feature_map, #Teniendo en cuenta este entrelazamiento
    ansatz=ansatz, # Actualizando los pesos del algoritmo
    loss="cross_entropy",
    optimizer=SPSA(maxiter=100), # Optimizando mediante SPSA en 100 iteraciones (el SPSA puede evaluar más de una vez por iteración)
    quantum_instance=quantum_instance, # Corriendo sobre la instancia de simulador
    callback=callback_graph # Obtenga y grafique los resultados en cada iteración
)
```

```python
# Codificación One-hot de las etiquetas (Variable de interés) 
def one_hot(train_labels):
   
    training_labels = list(train_labels) 
    train_labels = []
    
    for i in training_labels:
        j = []
        if i == 0:
            j.append(1) 
            j.append(i)
        else:
            j.append(0) 
            j.append(i)
        train_labels.append(j) 
    train_labels = np.array(train_labels)
                        
    return train_labels
```

```python
# Se codifican los datos en one-hot para realizar el entrenamiento del algoritmo de QML
Y_train=one_hot(Y_train)
```

```python

objective_func_vals = []
plt.rcParams["figure.figsize"] = (12, 6)

# Ajustar el Clasificador a los datos
#vqc.fit(X, y_one_hot) Formato de entrada de los datos
vqc.fit(X_train, Y_train)

# return to default figsize
plt.rcParams["figure.figsize"] = (6, 4)
```

```python
# Comprobación del entrenamiento
score = vqc.score(X_test,one_hot(Y_test))
print(f"Testing accuracy: {score:0.2f}")
```

```python
#Creación de matriz de confusion a partir de los datos reales y las predicciones del algoritmo
def one_hot_decoding(test_labels, predictions):
        
    predictions = list (predictions)
    test_labels = list (test_labels)
    truth = []
    predicted = []
    n = len (test_labels)
    
    for i in range(n):
        if test_labels[i] [0] == 1:
            truth.append(0)
        else: 
            truth.append (1)
            
        if predictions [1] [0] == 1: 
            predicted.append (0)
        else: 
            predicted.append (1)
    return truth, predicted
```

```python
from sklearn.metrics import confusion_matrix
#Predicciones vs. Datos reales
predictions = vqc.predict(X_train)
k, j = one_hot_decoding(Y_train, predictions)
confusion_matrix(k, j)
```
